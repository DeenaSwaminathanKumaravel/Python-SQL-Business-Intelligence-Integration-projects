## Day 07 - Hangman - Classic Word Game

```python
hangman()
```

```python
import random as rd

def hangman():
    stages = ['''
      +---+
      |   |
      O   |
     /|\  |
     / \  |
          |
    =========
    ''', '''
      +---+
      |   |
      O   |
     /|\  |
     /    |
          |
    =========
    ''', '''
      +---+
      |   |
      O   |
     /|\  |
          |
          |
    =========
    ''', '''
      +---+
      |   |
      O   |
     /|   |
          |
          |
    =========''', '''
      +---+
      |   |
      O   |
      |   |
          |
          |
    =========
    ''', '''
      +---+
      |   |
      O   |
          |
          |
          |
    =========
    ''', '''
      +---+
      |   |
          |
          |
          |
          |
    =========
    ''']
    word_list = ["aardvark", "baboon", "camel"]
    chosen_word = rd.choice(word_list)

    dashes = (["_"] * len(chosen_word))
    #print(chosen_word)
    
    print("Hangman - Guess the word!")
    print(dashes)
    rem_guess=6

    while rem_guess >= 0:
        guess = input(f"Guess the letter, you have {rem_guess} extra life remaining!").lower()

        for index, word in enumerate(chosen_word):
            if word == guess:
                dashes[index] = word

        if "_" not in dashes:
            print("You Won")
            break

        if guess not in chosen_word:
            print(stages[rem_guess])
            rem_guess-=1

            if rem_guess<0:
                print("You Lost")

    print(dashes)
```
